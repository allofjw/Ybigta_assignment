Grouping 이란? 
비지도학습은, Unlabeled data를 다루는 머신러닝기법에 대하여 논합니다. 
labeled된 데이터와 비교하여, 데이터로부터 패턴 형태를 찾아야한다는 특징이 존재하고 이를 Grouping이라고 합니다. Grouping은 크게, Clustering(군집화), Dimensionality Reduction(차원 축소) 두가지 분류가 존재하는데, 군집화의 대표적인 방법론은 K-means Clustering이 차원축소는 PCA가 있습니다.

K-means clustering은 centroid라는 중심점을 기준으로 k개의 군집화 개수를 설정하며 Grouping을 수행합니다. 이때 군집의 성격은 기계가 임의로 결정합니다. 즉, 사용자는 조절이 불가합니다.

PCA는 서로 의존성 높은 피쳐들을 함께 학습시킬 때 발생하는 '다중 공손성' 문제를 해결하여 피쳐들을 새로운 작은 차원으로 축소시키는 방법입니다.
PCA는 공분산 행렬에서 구한 고유벡터와 고유값을 선형변환하는 과정을 통해 수행할 수 있다. 






## 과제 2

```python
# 필요한 라이브러리 로드
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
# Train, Test 데이터 분리
X=df.iloc[:,:-1]
y=df.iloc[:,-1]

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
# 모델 학습
dt=DecisionTreeClassifier(random_state=10)
dt.fit(X_train,y_train)

# 모델 예측
y_pred=dt.predict(X_test)
```

```python
# 모델 평가
dt.score(X_test,y_test)

import sklearn.metrics as mt


accuracy = mt.accuracy_score(y_test, y_pred)
recall = mt.recall_score(y_test, y_pred)
precision = mt.precision_score(y_test, y_pred)
f1_score = mt.f1_score(y_test, y_pred)
matrix = mt.confusion_matrix(y_test, y_pred)

# 모델 평가 결과 출력
print('정확도 :', accuracy)
print('재현율 :', recall)
print('정밀도 :', precision)
print('f1 score :', f1_score)
print('confusion matrix :\n', matrix)
```
